```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataTrain = pd.read_csv("train.csv")
dataTest = pd.read_csv("test.csv")
dataTrain.head()
dataTest.head()
```

```python
categorical_vars = ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch',
       'Embarked', 'Pclass']


num_plots = len(categorical_vars)
total_cols = 8

if num_plots%total_cols==0:
    total_rows = num_plots//total_cols
else:
    total_rows = num_plots//total_cols + 1

fig, axs = plt.subplots(nrows=total_rows, ncols=total_cols,
                        figsize=(7*total_cols, 7*total_rows), constrained_layout=True, squeeze=False)

for i, var in enumerate(categorical_vars):
    row = i//total_cols 
    pos = i % total_cols 

    plot = sns.countplot(x=var, data=dataTrain, ax=axs[row][pos])
```

```python
def SubplotsVisualizacionVariables(categorical_vars, total_cols):
  
    num_plots = len(categorical_vars)
    if num_plots%total_cols==0:
        total_rows = num_plots//total_cols
    else:
        total_rows = num_plots//total_cols + 1

    fig, axs = plt.subplots(nrows=total_rows, ncols=total_cols,
                            figsize=(7*total_cols, 7*total_rows), constrained_layout=True, squeeze=False)

    for i, var in enumerate(categorical_vars):
        row = i//total_cols 
        pos = i % total_cols 

        plot = sns.countplot(x=var, data=dataTrain, ax=axs[row][pos])

    return
```

```python
SubplotsVisualizacionVariables(categorical_vars, 4)
```
